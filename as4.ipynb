## grph

```python
from Bio import SeqIO

l = []
id_list = []

with open('rosalind_grph.txt') as handle:
    for record in SeqIO.parse(handle, "fasta"):
        l.append(str(record.seq))
        id_list.append(str(record.id))

DG = nx.DiGraph()
k = 3

for i in range(0,len(l)):
    DG.add_node(id_list[i])
    DG.nodes[id_list[i]]['dna'] = l[i]

for node1 in DG.nodes:
    for node2 in DG.nodes:
        if node1 == node2:
            continue
        if DG.nodes[node1]["dna"][-3::] == DG.nodes[node2]["dna"][0:3]:
            DG.add_edge(node1, node2)

for node1 in DG.nodes:
    for node2 in DG.successors(node1):
        out = node1 + ' ' + node2
        print(out)
```

## tree

```python
def DFS(G, node):
    G.nodes[node]["visited"] = True
    if G.neighbors(node) == []:
        return
    for node2 in G.neighbors(node):
        if G.nodes[node2]["visited"] == False:
            DFS(G, node2)

filein = open('rosalind_tree.txt', encoding = 'UTF-8')
l = []
n = int(filein.readline())

G = nx.Graph()
nodes = np.arange(1, n + 1)
G.add_nodes_from(nodes)
edges = []

for line in filein:
    line = line.replace('\n', '')
    nodes = line.split(' ')
    edge =  (int(nodes[0]), int(nodes[1]))
    edges.append(edge)
G.add_edges_from(edges)
for node in G.nodes:
    G.nodes[node]["visited"] = False

i = 0
for node in G.nodes:
    if G.nodes[node]["visited"] == False:
        i = i + 1
        DFS(G, node)

print(i-1)
filein.close()
```

## long 

```python
from Bio import SeqIO

def final(s1, s2):
    combine2 = []
    overlap2 = []
    for i in range(len(s1)):
        if s1[i:] == s2[:len(s1)-i]:
            overlap2.append(s1[i:])
            combine2.append(s1+s2[len(s1)-i:])
            break
    for i in range(len(s2)):
        if s2[i:] == s1[:len(s2)-i]:
            overlap2.append(s2[i:])
            combine2.append(s2+s1[len(s2)-i:])
            break
    if not overlap2:
        return "", ""

    combine1 = min(combine2, key=len)
    overlap1 = max(overlap2, key=len)
    return combine1, overlap1

def ss(strings):
    temp = strings

    while len(temp) > 1:
        string = ""
        stringp = []
        length = 0

        for i in range(len(temp)-1):
            for j in range(i+1, len(temp)):
                combine1, overlap1 = final(temp[i], temp[j])
                if len(overlap1) > length:
                    string = combine1
                    stringp = [temp[i], temp[j]]
                    length = len(overlap1)

        temp.remove(stringp[0])
        temp.remove(stringp[1])
        temp.append(string)
        
    return temp 

l = []
id_list = []

with open('rosalind_long.txt') as handle:
    for record in SeqIO.parse(handle, "fasta"):
        l.append(str(record.seq))
    
result = ss(l)
print(result[0])
```

## bfs

```python
def BFS(G, node):
    L = [[]]
    L[0].append(node)
    G.nodes[node]["visited"] = True
    G.nodes[node]["distance"] = 0
    i = 0
    while L[i] != []:
        L.append([])
        for node1 in L[i]:
            for node2 in G.successors(node1):
                if G.nodes[node2]["visited"] == False:
                    G.nodes[node2]["visited"] = True
                    G.nodes[node2]["distance"] = i + 1 
                    L[i+1].append(node2)
        i = i + 1


filein = open('rosalind_bfs.txt', encoding = 'UTF-8')
n = int(filein.readline().split(' ')[0])

G = nx.DiGraph()
nodes = np.arange(1, n + 1)
G.add_nodes_from(nodes)
edges = []

for line in filein:
    line = line.replace('\n', '')
    nodes = line.split(' ')
    edge =  (int(nodes[0]), int(nodes[1]))
    edges.append(edge)
G.add_edges_from(edges)

for node in G.nodes:
    G.nodes[node]["visited"] = False
    G.nodes[node]["distance"] = -1

BFS(G, 1)

out = ''
for node in G.nodes:
    out = out + str(G.nodes[node]["distance"]) + ' '

out = out[0:len(out)-1]
filein.close()
print(out)
```

## dij

```python
filein = open('rosalind_dij.txt', encoding = 'UTF-8')
n = int(filein.readline().split(' ')[0])

G = nx.DiGraph()
nodes = np.arange(1, n + 1)
G.add_nodes_from(nodes)
edges = []

for line in filein:
    line = line.replace('\n', '')
    nodes = line.split(' ')
    edge =  (int(nodes[0]), int(nodes[1]), int(nodes[2]))
    edges.append(edge)
G.add_weighted_edges_from(edges)

for node in G.nodes:
    G.nodes[node]["visited"] = False
    G.nodes[node]["distance"] = -1

D = [9999] * n
D[0] = 0
res = [0] * n

temp = False

while not temp:
    temp = True
    for node in G.nodes:
        if G.nodes[node]["visited"] == False:
            temp = False
    if temp == True:
        break
    u = D.index(min(D)) + 1
    ind = D.index(min(D))
    val = min(D)
    G.nodes[u]["visited"] = True
    for v in G.successors(u):
        w = G[u][v]['weight']
        if G.nodes[v]["visited"] == True: 
            continue
        if D[u-1] + w < D[v-1]:
            D[v-1] = D[u-1] + w
    res[ind] = val
    D[ind] = 99999 #needed to not choose this node as u again

out = ''
for elem in res:
    out = out + str(elem) + ' '

out = out[0:len(out)-1].replace('9999', '-1') #9999 implies not reachable from Node 1
filein.close()
print(out)
```

## dag 

```python
def isAcyclic(G, node):
    G.nodes[node]["visited"] = 'visiting'
    if list(G.successors(node)) == []:
        G.nodes[node]["visited"] = True
        return 1
    for node2 in list(G.successors(node)):
        if G.nodes[node2]["visited"] != 'visiting':
            ret = isAcyclic(G, node2)
            G.nodes[node2]["visited"] = True
            if ret == -1:
                return -1
        else:
            G.nodes[node]["visited"] = True
            return -1
    G.nodes[node]["visited"] = True
    return 1

filein = open('rosalind_dag.txt', encoding = 'UTF-8')
k = int(filein.readline())

found = False
i = -1
graph_list = []
for line in filein:
    if line == '\n':
        found = True
        i = i + 1
        G = nx.DiGraph()
        continue
    if found:
        line = line.replace('\n', '')
        n = int(line.split(' ')[0])
        nodes = np.arange(1, n + 1)
        G.add_nodes_from(nodes)
        for node in G.nodes:
            G.nodes[node]["visited"] = False
            G.nodes[node]["parents"] = []
        graph_list.append(G)
        found = False
        continue  
    line = line.replace('\n', '')
    nodes = line.split(' ')
    graph_list[i].add_edge(int(nodes[0]), int(nodes[1]))

out = ''
res = []
r = 0
for G in graph_list:
    for node in G.nodes:
        r = isAcyclic(G, node)
        if r == -1:
            res.append('-1')
            break
    if r == 1:
        res.append('1')

for elem in res:
    out = out + elem + ' '
out = out[0:len(out)-1]

print(out)
filein.close()
```

## bip

```python
def isBip(G, node):
    L = [[]]
    G.nodes[node]["visited"] = True
    G.nodes[node]["distance"] = 0
    L[0].append(node)
    i = 0
    while L[i] != []:
        L.append([])
        for node1 in L[i]:
            color = G.nodes[node1]["color"]
            for node2 in G.neighbors(node1):
                if G.nodes[node2]["color"] == -1:
                    G.nodes[node2]["visited"] = True
                    G.nodes[node2]["distance"] = i + 1 
                    G.nodes[node2]["color"] = 1 if color == 0 else 0
                    L[i+1].append(node2)
                elif G.nodes[node2]["color"] == G.nodes[node1]["color"]:
                    return -1
        i = i + 1
    return 1

filein = open('rosalind_bip.txt', encoding = 'UTF-8')
k = int(filein.readline())

found = False
i = -1
graph_list = []
for line in filein:
    if line == '\n':
        found = True
        i = i + 1
        G = nx.Graph()
        continue
    if found:
        line = line.replace('\n', '')
        n = int(line.split(' ')[0])
        nodes = np.arange(1, n + 1)
        G.add_nodes_from(nodes)
        for node in G.nodes:
            G.nodes[node]["visited"] = False
            G.nodes[node]["color"] = -1
        graph_list.append(G)
        found = False
        continue  
    line = line.replace('\n', '')
    nodes = line.split(' ')
    graph_list[i].add_edge(int(nodes[0]), int(nodes[1]))

out = ''
res = []

for G in graph_list:
    G.nodes[1]["color"] = 0
    res.append(isBip(G, 1))

for elem in res:
    out = out + str(elem) + ' '
out = out[0:len(out)-1]

print(out)
filein.close()
```

## bf

```python
filein = open('rosalind_bf.txt', encoding = 'UTF-8')
n = int(filein.readline().split(' ')[0])

G = nx.DiGraph()
nodes = np.arange(1, n + 1)
G.add_nodes_from(nodes)
edges = []

for line in filein:
    line = line.replace('\n', '')
    nodes = line.split(' ')
    edge =  (int(nodes[0]), int(nodes[1]), int(nodes[2]))
    edges.append(edge)
G.add_weighted_edges_from(edges)

for node in G.nodes:
    if node == 1:
        G.nodes[node]["distance"] = 0
    else:
        G.nodes[node]["distance"] = float('inf')
        
for i in range(1, n):
    change = False
    for edge in G.edges:
        u = edge[0]
        z = edge[1]
        r = G.nodes[u]["distance"] + G[u][z]["weight"]
        if r < G.nodes[z]["distance"]:
            G.nodes[z]["distance"] = r
            change = True
    if not change:
        break

out = ''
res = [''] * n
for node in G.nodes:
    if G.nodes[node]["distance"] == float('inf'):
        G.nodes[node]["distance"] = 'x'
    res[node-1] = G.nodes[node]["distance"]
    
for elem in res:
    out = out + str(elem) + ' '
out = out[0:len(out)-1]

print(out)
filein.close()
```

## cte 

```python
filein = open('rosalind_cte.txt', encoding = 'UTF-8')
k = int(filein.readline())
initial_node_list= []

found = False
i = -1
graph_list = []

for line in filein:
    line = line.replace('\n', '')
    nodes = line.split(' ')
    if len(nodes) == 2:
        found = True
        i = i + 1
        G = nx.DiGraph()
        n = int(line.split(' ')[0])
        nodes = np.arange(1, n + 1)
        G.add_nodes_from(nodes)
        for node in G.nodes:
            G.nodes[node]["visited"] = False
            G.nodes[node]["cycle"] = False
        graph_list.append(G)
        continue  
    if found:
        initial_node_list.append((int(nodes[0]), int(nodes[1])))
        found = False
    graph_list[i].add_edge(int(nodes[0]), int(nodes[1]), weight = int(nodes[2]))

    
res = []
for i in range(0, len(graph_list)):
    s = initial_node_list[i][1]
    n = len(G.nodes)
    G = graph_list[i]
    for node in G.nodes:
        if node == s:
            G.nodes[node]["distance"] = 0
        else:
            G.nodes[node]["distance"] = float('inf')
            
    for j in range(1, n):
        change = False
        for edge in G.edges:
            u = edge[0]
            z = edge[1]
            r = G.nodes[u]["distance"] + G[u][z]["weight"]
            if r < G.nodes[z]["distance"]:
                G.nodes[z]["distance"] = r
                change = True
        if not change:
            break
            
    if G.nodes[initial_node_list[i][0]]["distance"] == float('inf'):
        res.append(-1)
    else:
        res.append(G.nodes[initial_node_list[i][0]]["distance"] + G[initial_node_list[i][0]][s]["weight"])
    

out = ''
for elem in res:
    out = out + str(elem) + ' '
out = out[0:len(out)-1]

print(out)
filein.close()
```

## nwc 

```python
filein = open('rosalind_nwc.txt', encoding = 'UTF-8')
k = int(filein.readline())
initial_node_list= []

found = False
i = -1
graph_list = []
for line in filein:
    line = line.replace('\n', '')
    nodes = line.split(' ')
    if len(nodes) == 2:
        found = True
        i = i + 1
        G = nx.DiGraph()
        n = int(line.split(' ')[0])
        nodes = np.arange(1, n + 1)
        G.add_nodes_from(nodes)
        graph_list.append(G)
        continue  
    if found:
        initial_node_list.append((int(nodes[0]), int(nodes[1])))
        found = False
    graph_list[i].add_edge(int(nodes[0]), int(nodes[1]), weight = int(nodes[2]))
        
for G in graph_list:
    s = 0
    G.add_node(s) #it's a node placed so that we can surely reach the loop, as this
                  #node is connected to every other node in the graph
    for node in G.nodes:
        if node != s:
            G.add_edge(s, node, weight = 0) # weight as 0 so it doesn't interfere with the BF algorithm
            G.nodes[node]["distance"] = float('inf')
        else:
            G.nodes[node]["distance"] = 0
                
res = []
for G in graph_list:
    cycle = False
    n = len(G.nodes)
    for i in range(1, n+1):
        if cycle:
            break
        change = False
        for edge in G.edges:
            u = edge[0]
            z = edge[1]
            r = G.nodes[u]["distance"] + G[u][z]["weight"]
            if r < G.nodes[z]["distance"]:
                G.nodes[z]["distance"] = r
                change = True
                if i == n:
                    cycle = True
                    break
        if not change:
            break
    if cycle:
        res.append(1)
    else:
        res.append(-1)
out = ''
for elem in res:
    out = out + str(elem) + ' '
out = out[0:len(out)-1]

print(out)
filein.close()
```

## sdag

```python
def DFS(G, node, ordered_graph):
    G.nodes[node]["visited"] = True
    for node2 in G.neighbors(node):
        if G.nodes[node2]["visited"] == False:
            DFS(G, node2, ordered_graph)
    ordered_graph.append(node)

filein = open('rosalind_sdag.txt', encoding = 'UTF-8')
n = int(filein.readline().split(' ')[0])

G = nx.DiGraph()
nodes = np.arange(1, n + 1)
G.add_nodes_from(nodes)
edges = []

for line in filein:
    line = line.replace('\n', '')
    nodes = line.split(' ')
    edge =  (int(nodes[0]), int(nodes[1]), int(nodes[2]))
    edges.append(edge)
G.add_weighted_edges_from(edges)

for node in G.nodes:
    G.nodes[node]["visited"] = False

ordered_graph = []
for node in G.nodes:
    if G.nodes[node]["visited"] == True:
        continue
    l = len(list(G.predecessors(node)))
    if l == 0:
        DFS(G, node, ordered_graph)    
ordered_graph.reverse()

for node in G.nodes:
    if node == 1:
        G.nodes[node]["distance"] = 0
    else:
        G.nodes[node]["distance"] = float('inf')
        
for u in ordered_graph:
    for v in G.successors(u):
        r = G.nodes[u]["distance"] + G[u][v]["weight"]
        if r < G.nodes[v]["distance"]:
            G.nodes[v]["distance"] = r

res = [''] * n
for node in G.nodes:
    if G.nodes[node]["distance"] == float('inf'):
        G.nodes[node]["distance"] = 'x'
    res[node-1] = G.nodes[node]["distance"]

out = ''
for elem in res:
    out = out + str(elem) + ' '
out = out[0:len(out)-1]

print(out)
filein.close()
```
